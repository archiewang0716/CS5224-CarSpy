## Processing the dataset

```python
import uuid
def uuid3(data):
    return (uuid.uuid3(uuid.NAMESPACE_DNS, str(data["model"])+"-"+str(data["year"])+"-"+str(data["price"])+"-"+str(data["mileage"])+"-"+str(data["fuelType"])+"-"+str(data["mpg"])+"-"+str(data["engineSize"])))
```

```python
import os
import string
import pandas as pd
directory = os.walk(r"archive")

data_merged = pd.DataFrame()
for path,dir_list,file_list in directory:  
    for file_name in file_list:
        file_path = os.path.join(path, file_name)
        data = pd.read_csv(file_path, sep=",")
        data.insert(0, 'brand', file_name[:-4])
        data_merged = pd.concat([data_merged,data],axis=0)
        data_merged = data_merged.reset_index(drop=True)
data_merged["uuid"] = data_merged.apply(uuid3, axis=1)
first_column = data_merged.pop('uuid')
data_merged.insert(0, 'uuid', first_column)

data_merged = data_merged.drop(['tax'],axis=1)
data_merged = data_merged.drop(['tax(Â£)'],axis=1)
data_merged["model"] = data_merged["model"].apply(lambda x: x.strip())
data_merged.insert(data_merged.shape[1], 'viewed', 0)
data_merged.insert(data_merged.shape[1], 'stared', 0)
print(data_merged)
data_merged.to_csv("used_cars.csv", sep=",", index=None)
```

```python
data_dummies = data_merged.copy()
data_dummies["brand_model"] = data_merged["brand"] + data_merged["model"]

data_dummies = data_dummies.drop("model", axis=1)
data_dummies = data_dummies.drop("brand", axis=1)
print(data_merged)
data_dummies = pd.get_dummies(data_dummies, columns=["brand_model", "transmission", "fuelType"])
data_dummies = data_dummies.drop("uuid", axis=1)
data_dummies
```

## AutoEncoder Recommendation

```python
from tensorflow.keras import Model
from tensorflow.keras.layers import Input, Dense
def encoder(x):
    x = Dense(units=128, activation='relu')(x)
    x = Dense(units=64, activation='relu')(x)
    return x

def decoder(x):
    x = Dense(units=128, activation='relu')(x)
    x = Dense(units=10, activation='sigmoid')(x)
    return x
x = Input(shape=(10,))
z = encoder(x)
x_recon = decoder(z)

autoencoder = Model(inputs=x, outputs=x_recon)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```
